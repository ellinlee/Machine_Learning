#convert to datetimes

```python
# Convert to datetimes
for key in df_dict.keys():
    tmp = df_dict[key]['time_iso'].astype(str) + " " + df_dict[key]['time_utc'].astype(str)
    timestamp = pd.to_datetime(tmp, format='%Y-%m-%d %H:%M:%S')
    df_dict[key].insert(0, 'timestamp', timestamp)
    df_dict[key].drop(columns=['time_iso', 'time_utc'], axis=1, inplace=True)

    #create new columns:month,weekday
    df_dict[key]=df_dict[key].assign(month=lambda x:x.timestamp.dt.month)
    df_dict[key].month=df_dict[key].month.astype('category')
    df_dict[key]=df_dict[key].assign(weekday=lambda x:x.timestamp.dt.weekday)
    df_dict[key].weekday=df_dict[key].weekday.astype('category')
    
    #print(df_dict[key])
    print(key)
    print(type(key))
```

#synchronize

```python
#output dictionary
dict_harmony={}

# Harmonize time series
for key in df_dict.keys(): 
    this_df = df_dict[key]
    this_df.index = this_df['timestamp']
    df_mean = this_df.resample('H').mean()
    df_max = this_df.resample('H').max()
    df_min = this_df.resample('H').min()
    
    # Create df_new by merging
    df_new = pd.merge(df_mean, df_max, left_index=True, right_index=True)
    df_new = pd.merge(df_new, df_min, left_index=True, right_index=True)
    
    # Reset index to column
    df_new.reset_index(inplace=True)
    
    # Append to new dictionary by update
    dict_harmony.update({key: df_new})
    
    print(dict_harmony[key])
```
