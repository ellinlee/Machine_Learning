# part 1. fill the missing values

```python
# Fill the data for May 3, 2022, to May 4, 2022 with mean values from May 2, 2022 and May 5, 2022

# Find the mean temperature, humidity, and light values for May 2, 2022
condition_1 = df_dict['df_gw_dachgeschoss']['time_iso'] == '2022-05-02'
meantmp_02 = df_dict['df_gw_dachgeschoss'][condition_1]['temperature'].mean()
meanhmd_02 = df_dict['df_gw_dachgeschoss'][condition_1]['humidity'].mean()
meanlight_02 = df_dict['df_gw_dachgeschoss'][condition_1]['light'].mean()

# Round the mean values to the nearest integer
meantmp_02=round(meantmp_02)
meanhmd_02=round(meanhmd_02)
meanlight_02=round(meanlight_02)
```

```python
# Find the mean temperature, humidity, and light values for May 5, 2022
condition_2=df_dict['df_gw_dachgeschoss']['time_iso']=='2022-05-05'
meantmp_05=df_dict['df_gw_dachgeschoss'][condition_2]['temperature'].mean()
meanhmd_05=df_dict['df_gw_dachgeschoss'][condition_2]['humidity'].mean()
meanlight_05=df_dict['df_gw_dachgeschoss'][condition_2]['light'].mean()

# Round the mean values to the nearest integer
meantmp_05=round(meantmp_05)
meanhmd_05=round(meanhmd_05)
meanlight_05=round(meanlight_05)
```

```python
# Calculate the mean temperature, humidity, and light values for May 2, 2022 and May 5, 2022
meantmp_final = (meantmp_02 + meantmp_05) / 2
meanhmd_final = (meanhmd_02 + meanhmd_05) / 2
meanlight_final = (meanlight_02 + meanlight_05) / 2

# Print the mean values
print(meantmp_final)
print(meanhmd_final)
print(meanlight_final)

# Create new rows for May 3, 2022 and May 4, 2022 with the calculated mean values
new_row_03 = pd.DataFrame({'time_iso': ['2022-05-03'], 'time_utc': ['00:00:00'], 'temperature': meantmp_final, 
                           'humidity': meanhmd_final, 'light': meanlight_final})
new_row_04 = pd.DataFrame({'time_iso': ['2022-05-04'], 'time_utc': ['00:00:00'], 'temperature': meantmp_final, 
                           'humidity': meanhmd_final, 'light': meanlight_final})
```

```python
#Convert new_row_03 to a DataFrame
new_row_03 = pd.DataFrame(new_row_03)

#Append new_row_04 to new_row_03
new_row_03 = new_row_03.append(new_row_04)

#Display the new_row_03 DataFrame to confirm
new_row_03
```

```python
# Convert the 'time_iso' column in new_row_03 to datetime format
new_row_03['time_iso'] = pd.to_datetime(new_row_03['time_iso'])

# Convert the 'humidity' column in new_row_03 to int64 data type
new_row_03['humidity'] = new_row_03['humidity'].astype('int64')

# Convert the 'light' column in new_row_03 to int64 data type
new_row_03['light'] = new_row_03['light'].astype('int64')

# Display information about the new_row_03 DataFrame
new_row_03.info()
```

```python
# Select rows from df_gw_dachgeschoss where the 'time_iso' column is less than '2022-05-05'
df1 = df_dict['df_gw_dachgeschoss'].loc[df_dict['df_gw_dachgeschoss']['time_iso'] < '2022-05-05']

# Select rows from df_gw_dachgeschoss where the 'time_iso' column is greater than or equal to '2022-05-05'
df2 = df_dict['df_gw_dachgeschoss'].loc[df_dict['df_gw_dachgeschoss']['time_iso'] >= '2022-05-05']
```

```python
#Display information about the df1 DataFrame
df1.info()
```

```python
# Append the contents of new_row_03 to df1 
df1 = df1.append(new_row_03, ignore_index=True)

# Append the contents of df2 to df1 
df1 = df1.append(df2, ignore_index=True)
```

```python
# Update the 'df_gw_dachgeschoss' in df_dict with the contents of df1
df_dict['df_gw_dachgeschoss'] = df1
```

# part 2 fix the anomalies

```python
# Find the index positions of rows where the humidity value is 0 in df_gw_dachgeschoss to find the anomalies
index_Dachgeschoss = df_dict['df_gw_dachgeschoss'].loc[df_dict['df_gw_dachgeschoss']['humidity'] == 0].index

# Display the index positions
index_Dachgeschoss
```

```python
# Select a subset of data from df_gw_dachgeschoss for calculating the mean
dataForMean = df_dict['df_gw_dachgeschoss'][12190:12250]

# Remove the rows with humidity value of 0 from the selected data
dataForMean = dataForMean.drop(index_Dachgeschoss)

# Calculate the mean of the 'temperature' column
TemMean = dataForMean['temperature'].mean()

# Calculate the mean of the 'humidity' column
HumMean = dataForMean['humidity'].mean()
```

```python
# Print the calculated mean value for temperature
print(TemMean)

# Print the calculated mean value for humidity
print(HumMean)
```

```python
# Iterate over each index in index_Dachgeschoss
for x in index_Dachgeschoss:
    # Update the 'temperature' value at index x with the rounded TemMean value
    df_dict['df_gw_dachgeschoss']['temperature'].iloc[x] = round(TemMean, 1)
    
    # Update the 'humidity' value at index x with the rounded HumMean value
    df_dict['df_gw_dachgeschoss']['humidity'].iloc[x] = round(HumMean, 1)
    
    # Print the updated row at index x in df_gw_dachgeschoss
    print(df_dict['df_gw_dachgeschoss'].iloc[x])
```

```python
# Find the index positions of rows where the humidity value is still 0 in df_gw_dachgeschoss
df_dict['df_gw_dachgeschoss'].loc[df_dict['df_gw_dachgeschoss']['humidity'] == 0].index
# Check that the outliers have been successfully updated
```

# convert to datetimes

```python
# Create datetime

# Combine the 'time_iso' and 'time_utc' columns as strings
dummy = df_gw_dachgeschoss['time_iso'].astype(str) + " " + df_gw_dachgeschoss['time_utc'].astype(str)

# Use pd.to_datetime to convert the combined strings into datetime objects
tmp = pd.to_datetime(dummy, format='%Y-%m-%d %H:%M:%S')

# Store the resulting datetime values in the 'tmp'
tmp
```

```python
# Convert to datetimes
for key in df_dict.keys():
    tmp = df_dict[key]['time_iso'].astype(str) + " " + df_dict[key]['time_utc'].astype(str)
    timestamp = pd.to_datetime(tmp, format='%Y-%m-%d %H:%M:%S')
    df_dict[key].insert(0, 'timestamp', timestamp)
    df_dict[key].drop(columns=['time_iso', 'time_utc'], axis=1, inplace=True)

    #create new columns:month,weekday
    df_dict[key]=df_dict[key].assign(month=lambda x:x.timestamp.dt.month)
    df_dict[key].month=df_dict[key].month.astype('category')
    df_dict[key]=df_dict[key].assign(weekday=lambda x:x.timestamp.dt.weekday)
    df_dict[key].weekday=df_dict[key].weekday.astype('category')
    
    #print(df_dict[key])
    print(key)
    print(type(key))
```

#synchronize

```python
#output dictionary
dict_harmony={}

# Harmonize time series
for key in df_dict.keys(): 
    this_df = df_dict[key]
    this_df.index = this_df['timestamp']
    df_mean = this_df.resample('H').mean()
    df_max = this_df.resample('H').max()
    df_min = this_df.resample('H').min()
    
    # Create df_new by merging
    df_new = pd.merge(df_mean, df_max, left_index=True, right_index=True)
    df_new = pd.merge(df_new, df_min, left_index=True, right_index=True)
    
    # Reset index to column
    df_new.reset_index(inplace=True)
    
    # Append to new dictionary by update
    dict_harmony.update({key: df_new})
    
    print(dict_harmony[key])
```
